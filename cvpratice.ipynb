```python
har=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
cap = cv2.VideoCapture(0)

while True:
    src, img=cap.read()
    img=cv2.resize(img,(250,250))
    faces=har.detectMultiScale(img,1.5,5)
    for x ,y, w, h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
    cv2.imshow('output',img)
    if cv2.waitKey(2)==27:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        img = cv2.resize(img,(244,244))
        if img is not None:
            images.append(img)
    return images
```

```python
Mask_images=load_images_from_folder('Mask\with mask')
```

```python
NoMask_images=load_images_from_folder('Mask\without mask')
```
